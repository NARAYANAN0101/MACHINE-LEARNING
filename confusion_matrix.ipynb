```python
import pandas as pd
```

```python
datset = pd.read_csv('/content/loan_approval_dataset.csv')
```

```python
datset
```

```python
df_encoding=pd.get_dummies(datset,columns=['education','self_employed','loan_status'],dtype=int,drop_first=True)
```

```python
df_encoding
```

```python
df_encoding=df_encoding.rename(columns={'education_Not Graduate':'Not Graduate',	'self_employed_Yes':'self_empolyed','loan_status_Rejected':'loan_rejected'})
```

```python
df_encoding
```

```python
indepedent=df_encoding[['no_of_dependents','income_annum','loan_amount','loan_term','cibil_score','residential_assets_value','commercial_assets_value','luxury_assets_value','bank_asset_value','Not Graduate',     "self_empolyed"]]
dependent=df_encoding[['loan_rejected']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(indepedent,dependent,test_size=1/3,random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier()
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
y_pred
```

```python
    from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
    cm=confusion_matrix(y_test,y_pred)
    print(cm)
```

```python
    cm_display = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=classifier.classes_)
    cm_display.plot()
```
