```python
import pandas as pd
```

```python
dataset=pd.read_csv('/content/heart.csv')
```

```python
dataset
```

```python
dataset.info()
```

```python
dataset.columns
```

```python
dataset.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
dataset_sc=scaler.fit_transform(dataset)
```

```python
dataset_sc
```

```python
from sklearn.impute import KNNImputer
imputer=KNNImputer(n_neighbors=3)
dataset_imputed=imputer.fit_transform(dataset_sc)
```

```python
dataset_imputed
```

```python
dataset_imputed=scaler.inverse_transform(dataset_imputed)
dataset_imputed
```

```python
dataset_imputed=pd.DataFrame(dataset_imputed,columns=dataset.columns)
dataset_imputed .head()
```

```python
dataset_imp=dataset_imputed.round().astype(int)
```

```python
final_datset=pd.concat([dataset,dataset_imp],axis=1)
final_datset
```

```python
independent=final_datset[['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach',
       'exang', 'oldpeak', 'slope', 'ca', 'thal']]
dependent=final_datset[['target']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
classification=RandomForestClassifier()
classification.fit(x_train,y_train)
```

```python
y_pred=classification.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
cm=accuracy_score(y_test,y_pred)
```

```python
print("accuarcy Score:",cm)
```
