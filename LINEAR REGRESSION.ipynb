```python
import pandas as pd
```

```python
dataset=pd.read_csv("/content/loan_approval_dataset.csv")
```

```python
dataset
```

```python
df_encoding=pd.get_dummies(dataset,columns=["education","self_employed","loan_status"],dtype=int,drop_first=True)
```

```python
df_encoding
```

```python
name = df_encoding.rename(columns={'education_Not Graduate':'EDUCATION','self_employed_Yes':'SELF_EMPOLYMENT'})
```

```python
name = df_encoding.drop('loan_status_Rejected', axis=1)
name = name.rename(columns={'education_Not Graduate':'EDUCATION','self_employed_Yes':'SELF_EMPOLYMENT'})
```

```python
name
```

```python
from sklearn.linear_model import LinearRegression
```

```python
df=pd.DataFrame(name)
```

```python
df
```

```python
independent=df[["no_of_dependents","income_annum","loan_term","cibil_score","residential_assets_value","commercial_assets_value","luxury_assets_value"	,"bank_asset_value","EDUCATION","SELF_EMPOLYMENT"]]
dependent=df[["loan_amount"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=42)
```

```python
X_test
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
Y_test
```

```python
from sklearn.metrics import r2_score
```

```python
r2=r2_score(Y_test,y_pred)
```

```python
r2
```
