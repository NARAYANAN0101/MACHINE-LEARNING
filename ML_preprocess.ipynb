```python
final_datset=pd.concat([object_cols,imputed_df],axis=1)
```

```python
final_datset
```

```python
independent=final_datset.drop(["classification"], axis=1)
dependent=final_datset[["classification"]]
```

```python
from sklearn.model_selection import train_test_split
X_test,X_train,Y_test,Y_train=train_test_split(independent,dependent,test_size=0.30,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
classifications=KNeighborsClassifier()
classifications.fit(X_train,Y_train)
```

```python
y_pred=classifications.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
ACC=accuracy_score(y_pred[:,0],Y_test.iloc[:, 0])
print("Accuracy Score:",ACC)
```
